## ¿Cómo funciona el juego?
Vamos a realizar una versión que tiene algunas particularidades respecto al juego original, de manera que sea más sencillo el desarrollo. Veamos cómo funciona:

1. Hay dos jugadores: la máquina y tú
2. Un **tablero de 10 x 10** posiciones donde irán los barcos.
3. Lo primero que se hace es colocar los barcos. Para este juego **los barcos se colocan de manera aleatoria. Ahora bien, puedes empezar colocando los barcos en unas posiciones fijas, que no cambien con cada partida, y después implementarlo aleatoriamente, ya que es más complejo. Los barcos son:**
    * 4 barcos de 1 posición de eslora
    * 3 barcos de 2 posiciones de eslora
    * 2 barcos de 3 posiciones de eslora
    * 1 barco de 4 posiciones de eslora

4. Tanto la máquina como tú tenéis un tablero con barcos, y se trata de ir "disparando" y hundiendo los del adversario hasta que un jugador se queda sin barcos, y por tanto, pierde.
5. Funciona por turnos y empiezas tú.
6. En cada turno disparas a una coordenada (X, Y) del tablero adversario. **Si aciertas, te vuelve a tocar**. En caso contrario, le toca a la máquina.
7. En los turnos de la máquina, si acierta, también le vuelve a tocar. ¿Dónde dispara la máquina? A un punto aleatorio en tu tablero.
8. Si se hunden todos los barcos de un jugador, el juego acaba y gana el otro.
9. Por norma del juego no puede haber espacios entre barcos. Ignora esta norma

```python
import numpy as np

tablero = np.full((10, 10), " ")

def colocar_barco(tablero, fila_inicio, columna_inicio, longitud, orientacion):

    if orientacion == 'horizontal' and columna_inicio + longitud > 10:
        return "Barco mal colocado"
    elif orientacion == 'vertical' and fila_inicio + longitud > 10:
        return "Barco mal colocado"
        
    
    if orientacion == 'horizontal':
            if "O" in tablero[fila_inicio, columna_inicio:columna_inicio+longitud]:
                return "Barco mal colocado"
            else:
                tablero[fila_inicio, columna_inicio:columna_inicio+longitud] = "O"
    
    elif orientacion == 'vertical':
            if "O" in tablero[fila_inicio:fila_inicio+longitud, columna_inicio]:
                return "Barco mal colocado"
            else:
                tablero[fila_inicio:fila_inicio+longitud, columna_inicio] = "O"
                
    return "Barco colocado correctamente.\n"


barcos = [4, 3, 3, 2, 2, 2, 1, 1, 1, 1]  
seleccion = input("¿Desea una selección manual o aleatorio?")

if seleccion == "manual":
    print("Por favor, coloca tus barcos. Los barcos no pueden solaparse ni salirse del tablero.")

    for i in range(10):
        print(f"Colocando barco de tamaño {barcos[i]}")

        valores = list(input("Introduzca la fila,columna y orientación: ").split(","))
        
        fila_inicio = int(valores[0])
        columna_inicio = int(valores[1])
        orientacion = str(valores[2])
        
        resultado = colocar_barco(tablero, fila_inicio, columna_inicio, barcos[i], orientacion)
        
        while True:
            if resultado != "Barco colocado correctamente.\n":
                valores = list(input("Error al colocar. Introduzca coordenadas correctas: ").split(","))
                fila_inicio = int(valores[0])
                columna_inicio = int(valores[1])
                orientacion = str(valores[2])
                resultado = colocar_barco(tablero, fila_inicio, columna_inicio, barcos[i], orientacion)

            if resultado == "Barco colocado correctamente.\n":
                break
        print(resultado)

if seleccion == "aleatorio":
    for i in range(10):
        print(f"Colocando barco de tamaño {barcos[i]}")

        resultado = colocar_barco(tablero, np.random.choice(range(0,9,1)),
        np.random.choice(range(0,9,1)), barcos[i], 
        np.random.choice(["verical","horizontal"]))
        
        while True:
            if resultado == "Barco colocado correctamente.\n":
                break

            if resultado == "Barco mal colocado":
                fila_inicio = np.random.choice(range(0,9,1))
                columna_inicio = np.random.choice(range(0,9,1))
                orientacion = np.random.choice(["verical","horizontal"])
                resultado = colocar_barco(tablero, fila_inicio, columna_inicio, barcos[i], orientacion)           
        print(resultado)
else:
    print("Inserte primero si quiere una selección Manual o Aleatoria")
print(tablero)
```

```python
import numpy as np

class Barco:
    def __init__(self, tamaño):
        self.tamaño = tamaño

class Tablero:
    def __init__(self, tamaño):
        self.tamaño = tamaño
        self.tablero = np.full((tamaño, tamaño), ' ')
        self.barcos = []

    def añadir_barco(self, barco, fila, columna, orientacion):
        if orientacion not in ['horizontal', 'vertical']:
            raise ValueError("La orientación debe ser 'horizontal' o 'vertical'")
        if fila < 0 or fila >= self.tamaño or columna < 0 or columna >= self.tamaño:
            raise ValueError("Las coordenadas están fuera del tablero")
        if orientacion == 'horizontal' and columna + barco.tamaño > self.tamaño:
            raise ValueError("El barco no cabe en el tablero en la dirección especificada")
        if orientacion == 'vertical' and fila + barco.tamaño > self.tamaño:
            raise ValueError("El barco no cabe en el tablero en la dirección especificada")

        for i in range(barco.tamaño):
            if orientacion == 'horizontal':
                if self.tablero[fila, columna+i] != ' ':
                    raise ValueError("Ya hay un barco en esa posición")
            else:
                if self.tablero[fila+i, columna] != ' ':
                    raise ValueError("Ya hay un barco en esa posición")

        for i in range(barco.tamaño):
            if orientacion == 'horizontal':
                self.tablero[fila, columna+i] = 'B'
            else:
                self.tablero[fila+i, columna] = 'B'

        self.barcos.append(barco)

    def mostrar(self):
        print(self.tablero)

tamaños_barcos = [4, 3, 2, 2, 1, 1]
tablero = Tablero(10)

for tamaño in tamaños_barcos:
    while True:
        try:
            fila = int(input(f"Introduce la fila para el barco de tamaño {tamaño}: "))
            columna = int(input(f"Introduce la columna para el barco de tamaño {tamaño}: "))
            orientacion = input(f"Introduce la orientación (horizontal/vertical) para el barco de tamaño {tamaño}: ")
            barco = Barco(tamaño)
            tablero.añadir_barco(barco, fila, columna, orientacion)
            break
        except ValueError as e:
            print(e)

tablero.mostrar()
```
