```python
import numpy as np

#Tableros
tablerocpu_b = np.full((10,10)," ")
tablerocpu_t = np.full((10,10)," ")
tablerouser_b = np.full((10,10)," ")
tablerouser_t = np.full((10,10)," ")


#coloco barcos fijos CPU
tablerocpu_b[3,3:7] =  "O" #4esloras
tablerocpu_b[1:4,9] =  "O" #3esloras
tablerocpu_b[6:9,6] =  "O"  #3esloras
tablerocpu_b[8,1:3] =  "O" #2esloras
tablerocpu_b[0:1,1:3] = "O" #2esloras
tablerocpu_b[0:1,7:9] = "O" #2esloras
tablerocpu_b[6:7,1:3] = "O" #1esloras
tablerocpu_b[9,4] = "O" #1esloras
tablerocpu_b[2,1] = "O" #1esloras
tablerocpu_b[8,8] = "O" #1esloras

#barcos fijos user
tablerouser_b[4,5:9] =  "O" #4esloras
tablerouser_b[2,2:5] =  "O" #3esloras
tablerouser_b[6:9,7] =  "O" #3esloras
tablerouser_b[8,1:4] =  "O" #2esloras
tablerouser_b[0:1,1:3] = "O" #2esloras
tablerouser_b[0:1,7:9] = "O" #2esloras
tablerouser_b[6:7,1:3] = "O" #1esloras
tablerouser_b[9,6] = "O" #1esloras
tablerouser_b[4,1] = "O" #1esloras
tablerouser_b[9,9] = "O" #1esloras



#Inicio de Partida
barcoscpu = 20 #es la suma de todas las esloras, cuando las esloras se termine, se usará break o saldra del bucle
barcosuser = 20 #es la suma de todas las esloras, cuando las esloras se termine, se usará break o saldrá del bucle

while barcoscpu > 0 and barcosuser > 0:
    
    print("Tablero de barcos del CPU\n")
    print(tablerocpu_b,"\n")
    print("Tablero de tiros del CPU\n")
    print(tablerocpu_t,"\n")  
        
    print("Tablero de barcos del user\n")
    print(tablerouser_b,"\n")
    print("Tablero de tiros del user\n")
    print(tablerouser_t,"\n")
    
    
    #pido coordenadas y coloca en tupla
    tiroux = int(input("ingresa coordenada en x"))
    tirouy = int(input("ingresa coordenada en y"))
    
    #verifico objetivo de disparo de user
    if tablerocpu_b[tirouy,tiroux] == "X" or tablerocpu_b[tirouy,tiroux] == "F": #si la coordenada ya ha sido ingresada -user-, y fue Asertivo o Fallado -tablero cpu-
        print("Escoge otras coordenadas")
    elif tablerocpu_b[tirouy,tiroux] == "O": #tiro asertivo, 
        print("Tiro en el blanco, dispara de nuevo")
        barcoscpu = barcoscpu-1 #disminuyo total de esloras cpu
        tablerocpu_b[tirouy,tiroux] = "X" #modifico tablero barcos cpu
        tablerouser_t[tirouy,tiroux] = "X" #modifico trablero tiros user
    elif tablerocpu_b[tirouy,tiroux] == " ":
        tablerocpu_b[tirouy,tiroux] = "F" #modifico tablero barcos cpu
        tablerouser_t[tirouy,tiroux] = "F" #modifico trablero tiros user
        print("Fallaste, turno de la computadora")
        tirocpu = np.random.randint(10,size = (2,)) #tiro random computadora en arreglo de 1 dim
        while tablerouser_b[tirocpu[0],tirocpu[1]] == "X" or tablerouser_b[tirocpu[0],tirocpu[1]] == "F": #mientras tire en un tiro atinado o en un tiro fallado, genera otro
            tirocpu = np.random.randint(10,size = (2,))
        while tablerouser_b[tirocpu[0],tirocpu[1]] == "O": #tiro random cpu asertivo
            barcosuser = barcosuser - 1 #disminuyo total de esloras cpu
            tablerouser_b[tirocpu[0],tirocpu[1]] = "X" #cambio valor en tablero de barcouser tiro atinado X
            tablerocpu_t[tirocpu[0],tirocpu[1]] = "X" #cambio valor en tablero de barcouser tiro atinado X
            print("La CPU tiró asertivamente")
            if barcosuser == 0: #condicional para romper el programa ciclo si todas las esloras han sido impactadas.
                break
            tirocpu = np.random.randint(10,size = (2,)) #vuelve a tirar por haber atinado y si aun quedan esloras 
        tablerouser_b[tirocpu[0],tirocpu[1]] = "F" #cambio valor en tablero user a F
        tablerocpu_t[tirocpu[0],tirocpu[1]] = "F" #cambio valor en tablero user a F
        print("La CPU falló")

if barcoscpu == 0:
    print("\n\n El CPU te ha vencido")

elif barcosuser == 0:
    print("\n\n Has vencido al CPU")


        

```
